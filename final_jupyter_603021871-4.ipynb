# 2.Treemap

```python
import squarify 
```

```python
volume = [61000, 7878, 1791]
```

```python
squarify.plot(sizes=volume)
```

```python
labels = ['Confirmed\n volume: 61000', 'Recovered\n volume: 7878',    # labels คือ ชื่อของ4เหลี่ยม
          'Deaths\n volume: 1791']
color_list = ['#f9d4d4', '#d35158', '#ea3033']
```

```python
squarify.plot(sizes=volume, label=labels,
              color=color_list, alpha=1)   
```

```python
squarify.plot(sizes=volume, label=labels,
              color=color_list, alpha=0.7) 
```

# 3.Word cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np   
```

```python
df = pd.read_csv('C:\\Users\\Administrator\\Desktop\\Basicprogramming\\2019_nCoV_data.csv')
```

```python
df
```

```python
CoV_data = df.iloc[:,3]
```

```python
CoV_data
```

```python
CoV_data[0]
```

```python
text = ''
```

```python
for i in range(len(CoV_data)):
    text += CoV_data[i]
```

```python
text
```

```python
wordcloud = WordCloud(width=1280, height=853, margin=0,
                      colormap='Blues', font_step=4).generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.margins(x=0, y=0)
```
