### 1> สร้าง function ชื่อ smartDict ซึ่งทำหน้าที่:
-function รับ input เป็น list2 ตัวที่มีขนาดเท่ากันโดย list แรกเป็น list ของตัวเลข list ที่2เป็นlistของtext

-ถ้าค่าในlistแรก เป็นเลขคี่ ให้ค่าใน dict ที่เป็นoutput มีkeyเป็นค่าจากlistตัวที่1 และvalueจากlistตัวที่2

-ถ้าค่าในlistแรก เป็นเลขคู่ ให้ค่าใน dict ที่เป็นoutput มีkeyเป็นค่าจากlistตัวที่2 และvalueจากlistตัวที่1

-output ของfunction คือ dict ที่สร้างขึ้น


เช่น 
output1 = smartDict([1,2,3],['apple','banana','orange'])

จะได้ output1 เป็น ,{1:'apple','banana':2,3:'orange'}

```python
def smartDict( listnumber,listtext):
    if listnumber : 
        retdict = {} 
        for index in range(len(listnumber)):  
            retdict[listtext[index]]= listnumber[index] 
    else:
        retdict = {} 
        for index in range(len( listnumber)):  
             retdict[listnumber[index]]= listtext[index] 
            
    return retdict  
```

```python
output1 = smartDict([1,2,3],['apple','banana','orange'])
```

## 2> จากไฟล์Wuhan_ncoV_GNome.txt จงเขียนโปรแกรมเพื่อตอบคำถามต่อไปนี้

-กรดอะมอโนใด(A,T,G,C)มีปริมาณมากที่สุดและมีเท่าไหร่

-จงหา sequence ที่มีขนาด3และ4ที่เกิดขึ้นซ้ำมากที่สุด

```python
import pandas as pd
```

```python
DNAseq = pd.read_csv('C:\\Users\\Administrator\\Desktop\\Basicprogramming\\Wuhan_nCoV_GNome.txt',header=None)
```

```python
DNAseq.head()
```

```python
def countChar(seq,character):
    charcount = 0
    for char in seq:
         if char == character:
                charcount = charcount + 1
    return charcount
```

```python
countChar('ATTAAAGGTTTATACCTTCCCAGGTAACAAACCAACCAACTTTCGA','A')
```

```python
ANum = 0 
TNum = 0
GNum = 0 
CNum = 0 
for row in range(DNAseq.shape[0]):
    ANum += countChar(DNAseq.iloc[row][0],'A') 
    TNum += countChar(DNAseq.iloc[row][0],'T')
    GNum += countChar(DNAseq.iloc[row][0],'G')
    CNum += countChar(DNAseq.iloc[row][0],'C')
    
print(f'There are {ANum} As, {TNum} Ts, {GNum} Gs, {CNum} Cs')
```

#### ตอบ กรดอะมิโน T มีมากที่สุด จำนวน 9594 

```python
'ATTAAAGGTTT'
```

```python
def populate3charstring(seq):
    str_list = list()
    for i in range(len(seq)-2):
        str_list.append(seq[i:i+3])
    return str_list
```

```python
populate3charstring('ATTAAAGGTTT')
```

```python
list_all_3char = list()
for row in range(DNAseq.shape[0]):
    list_all_3char = list_all_3char + populate3charstring(DNAseq.iloc[row][0]) 
```

```python
len(list_all_3char)
```

```python
unique_3char = list(set(list_all_3char))
```

```python
len(unique_3char)
```

```python
dict_3char = {}
for sub_str in unique_3char:
    num = 0
    for s in list_all_3char:
        if sub_str == s:
            num += 1
    dict_3char[sub_str] = num
```

```python
dict_3char 
```

```python
maxVal = -100
maxKey = None
for k in dict_3char.keys():
    if dict_3char[k] > maxVal:
        maxVal = dict_3char[k]
        maxKey = k
```

```python
maxKey
```

```python
print(f'max 3-long-sequence is {maxKey} with frequency equals {maxKey}')
```

#### ตอบ max 3-long-sequence is TTT with frequency equals TTT
