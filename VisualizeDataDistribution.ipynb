```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
setosa_np[:5,:]
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],setosa_np[:,2])  #แกนxคือหลักที่1 แกนyคือหลักที่2 แกนที่3คือตัวที่3
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2])  #อยากให้จุดมันใหญ่ขึ้นในหลัง่ที3
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2],c = 'k') #พารามิเตอร์ตัวที่1,2,3,4   และcคือสี
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2],c = 'c') 
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2],c = 'r') 
```

```python
print(f'max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

```python
colors = [ ]
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i < 0.3:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else:
        colors.append('r')
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c = colors) 
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c = setosa_np[:,3],cmap='YlOrBr' )   #แบบไล่สี สีเข้มค่ามาก สีอ่อนค่าน้อย  #ใช้ตัวอักษรเฉพาะสี
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c = np.array([[1,0.5,0.8]]) )   #1ตัวแรกเปลี่ยนว่าเราใช้สีแดงเต็มๆเลย 1,และ2สีเขียวใส่0.5
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c = setosa_np[:,3],cmap='YlOrBr',marker="*" ) #พารามิเตอร์ตัวที่5
```
